# Vectors
## Create Array

```python
# creating vectors with np.arange
seq_numbers1 = np.arange(12)
print("s1 = ",seq_numbers1)

seq_numbers2 = np.arange(12,23)
print("s2 = ",seq_numbers2)

seq_numbers3 = np.arange(24,50,2.5)
print("s3 = ",seq_numbers3)

# use existing list to create array
list_of_nums = [4,8,15,16,23,42]
array = np.array(list_of_nums)

# check array format and type
print("shape: ", array.shape)
print("datatype:", array.dtype)
print("dimention:", array.ndim)
print("size: ", array.size)
print("itemsize: ", array.itemsize)
print("strides: ", array.strides)
print("flag: ", array.flags)
print("arraytype: ", type(array))
print("----np.info-------")
print(np.info(array))

# specify dtype
print("---specify dtype-------")
ar2 = np.array(list_of_nums,dtype = np.int8)
print("dtype: ",ar2.dtype)
```

```python
# special creating
print(np.zeros(5))
print(np.ones(5)*2)
print(np.array([np.nan,1,2]))

# boolean vector
bi_vectors = np.array([True, False, True, True, False, True])
print(bi_vectors)
print(bi_vectors.dtype)
```

```python
# get / set values are the same as python
array = np.array([4,8,15,16,23,42])

first_element = array[0]
last_element = array[-1]
print("First element: ", first_element)
print("Last element: ", last_element)
array[-1] = 100
print("changed value:", array[-1])
```

## Functions 
### Basic Operations

```python
fi = np.arange(1,5) # 1,2,3,4
se = np.array([6,8,3,2]) # 6,8,3,2

print("Sum: ",fi + se)
print("contraction: ",fi - se)
print("Multiple: ", fi * se)
print("division:",np.around(fi / se,2))
print("Exponentiation: ", fi ** se) 
print("add 100:", fi + 100)
```

### Aggragate functions

```python
array = np.array([6,8,3,2]) 
print("array",array)
print("Sum: ", np.sum(array))
print("Ave: ", np.mean(array))
print("Product: ", np.prod(array))
print("Max: ", np.max(array))
print("Min: ", np.min(array))
print("Argmax: ", np.argmax(array)) # return index
print("Argmin: ", np.argmin(array))


```

## Slicing


```python

ar1 = np.arange(5) # 0,1,2,3,4
print(ar1[3:])
print(ar1[82:]) # return an empty array
print("--------------------")

print(ar1[:3])
print(ar1[:82]) # return whole array
print(ar1[:-1])
print("--------------------")

print(ar1[1:3]) 
print(ar1[1:-1])
print(ar1[1:-1:2])

print("--------------------")

# reverse array
print(ar1[::-1])
```

### Views vs Copies

```python
%%time  
# check CPU speedtime
# Views: change happened on reverse array also impact on original array
array = np.arange(5)
reverse = array[::-1]
print("Original:", array)
reverse[0] = 1000
print("Modified: ",reverse)
print("Original: ",array)
print("View?",reverse.base is array) # check if array is the base of reverse
print("View?",np.shares_memory(reverse, array)) # check if they use same memory
```

```python
%%time  
# check CPU speedtime
#Copies： copies on new memory
array = np.arange(5)
reverse = np.copy(array[::-1])
print("Original:", array)
reverse[0] = 1000
print("Modified: ",reverse)
print("Original: ",array)
print("View?",reverse.base is array) # check if array is the base of reverse
print("View?",np.shares_memory(reverse, array)) # check if they use same memory
```

## Sorting

```python
array = np.array([16,4,42,8,15,23])

#return a sorted array
sorted_array = np.sort(array,kind = 'heapsort')
# this won't change original array
print("Origin array:")
print(array)
print("sorted array:")
print(sorted_array)

# in-place sorting 
array.sort() 
print("Origin array:")
print(array)
# this method won't return anything, don't get values from it like ar = array.sort() 
```
