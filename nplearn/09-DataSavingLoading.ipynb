```python
import numpy as np
```

# import txt file

```python
data = np.loadtxt("data.txt", delimiter=",")
print(data)
print(data.mean())
```

# import csv file

```python
data2 = np.loadtxt("data_csv.csv",delimiter=",")
print(data2)
print(np.isnan(data2).any())
print(data2.mean())
total_mean = np.nanmean(data2)
print(total_mean)
```

# save txt/csv file

```python
np.nan_to_num(data2,copy = False,nan = total_mean)
```

```python
np.savetxt("newdata.txt",data2,delimiter=",",fmt="%.2f",header ="comment") 
np.savetxt("newdata.csv",data2,delimiter=",",fmt="%.2f",header ="comment") 
```

# load and save numpy objects

```python
#save
daily_mean = data2.mean(axis = 1)
print(daily_mean)
np.save("daily_mean_new",daily_mean)
```

```python
#load
load_data =np.load("daily_mean_new.npy")
print(load_data)
```

## compress two array in one file

```python
#save file and name the array
np.savez("rainy_days",load_mean = load_data,raw_data = data2)
```

```python
#load
load_npz =np.load("rainy_days.npz")
print(load_npz.files)
load_mean = load_npz["load_mean"]
raw_data = load_npz["raw_data"]
```
