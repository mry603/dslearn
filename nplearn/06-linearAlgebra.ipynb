```python
import numpy as np
from numpy import linalg # package for liner Algebra
# The vectors we are going to use in this module.
first_vector = np.array([1, 2, 4])
second_vector = np.array([2, -1, 9])
# The matrices we are going to use in this module.
first_matrix = np.array([[1, 2, 0], [3, 10, -2], [-3, 5, 3]])
second_matrix = np.array([[1, 2, 0], [0, 0, -2], [-3, 8, 3]])
print("vectors:", first_vector,second_vector)
print("matrix:\n", first_matrix,"\n")
print(second_matrix)
```

# Product
## pointwise multiplication

```python
print("vector * vector: ", first_vector * second_vector)
print("matrix * matrix: ", first_matrix * second_matrix)
```

## Dot product
### vector

```python
print(first_vector)
print(second_vector)
np.dot(first_vector,second_vector)
```

```python
def are_orthogonal(vect1,vect2):
    return np.dot(vect1,vect2) ==0
```

### matrix

```python
print(first_matrix)
print(np.eye(3) @ first_matrix)
print(second_matrix)
print(np.dot(first_matrix,second_matrix))
print(first_matrix @ second_matrix)
print(np.matmul(first_matrix,second_matrix))
```

```python
# matrix power
print(first_matrix @ first_matrix @ first_matrix @ first_matrix)
print(linalg.matrix_power(first_matrix,4))
```

## Cross product

```python
fscross = np.cross(first_vector, second_vector)
print(fscross)
print(np.dot(first_vector,fscross))
print(np.dot(second_vector,fscross))
```

```python
vec1 = np.arange(2)
vec2 = np.arange(1,3)
print(vec1)
print(vec2)
print(np.cross(vec1,vec2))
```

# attributes from linalg
## norm

```python
# Norm
print("first_vector norm",linalg.norm(first_vector))
print("second vector norm",linalg.norm(second_vector))

print("first_matrix norm",linalg.norm(first_matrix))
```

```python
def angle(vec1,vec2):
    cos_angle = np.dot(vec1,vec2)/(linalg.norm(vec1) * linalg.norm(vec2))
    return np.arccos(cos_angle)
```

```python

print(angle(first_vector,second_vector))

print(angle(first_matrix,second_matrix))
```

## Determinant
if det is 0, it is not invertable. If det is non zero, it is invertable

```python
print(linalg.det(first_matrix))
print(linalg.det(second_matrix))

non_invertible = np.array([[1,2,0],[-2,10,3],[-3,8,3]])
print(linalg.det(non_invertible))
print(np.around(linalg.det(non_invertible)))
```

## Inverse

```python
print(np.around(linalg.inv(first_matrix) @ first_matrix,10)) # np.dot

print(linalg.inv(second_matrix))
```

```python
# Ax = b, x = A-1 * b
x1= linalg.solve(first_matrix,first_vector)
print(x1)

print(first_matrix @ x1) # equal to first_vector

# no solution
x2= linalg.solve(non_invertible,first_vector)
print(x2)

print(non_invertible @ x2) # not equal to first_vector

# infinate solution

third_vector = np.array([1,-2,-3])
x3= linalg.solve(non_invertible,third_vector)
print(x3)

print(non_invertible @ x3) # equal to third_vector but not the noly solution
```
