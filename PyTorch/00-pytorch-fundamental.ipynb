## range and steps of tensors

```python
one_to_ten= torch.arange(1,11)
step= torch.arange(start=0,end = 1, step = 0.15)

one_to_ten,step
```

## tensors like

```python
ten_zeros = torch.zeros_like(input = step)
ten_ones = torch.ones_like(input = step)
ten_empty = torch.empty_like(input = step)
ten_rand = torch.rand_like(input = step)
ten_randn = torch.randn_like(input = step)
ten_randint = torch.randint_like(input = step, low= 5, high= 10)
ten_full = torch.full_like(input = step,fill_value=7)

print("zeros_like:",ten_zeros)
print("ones_like:",ten_ones)
print("empty_like:",ten_empty)
print("rand_like:",ten_rand)
print("randn_like:",ten_randn)
print("randint_like:",ten_randint)
print("full_like:",ten_full)
```

## setting tensor datatypes

```python
float_32_tensor = torch.tensor([3.0,6,9],
                               dtype = None, # data type
                               device = None, # device type: cpu, cuda, mps, xpu
                               requires_grad=False) # Is True if gradients need to be computed for this Tensor, False otherwise.
print(float_32_tensor.dtype)
float_16_tensor = torch.tensor([1.0,6,9],dtype = torch.float16)
print(float_16_tensor.dtype)
float_16_tensor = float_32_tensor.type(torch.float16)
print(float_16_tensor.dtype)
print(float_16_tensor)
```

# Tensor oprations
## get attributes

```python
some_tensor = torch.rand(3,4)
print(some_tensor)
# get dtype
print("datatype:",some_tensor.dtype)
print("complex: ",some_tensor.is_complex())
print("floating: ",some_tensor.is_floating_point())

# shape
print("shape:",some_tensor.shape)

#device
print("device:",some_tensor.device)
print("cpu:",some_tensor.is_cpu)
```

## manipulating tensors

```python
tensor1 = torch.tensor([1,2,3])

print("addition:")
print(tensor1 + 10)
print(tensor1.add(10))
print(torch.add(tensor1,10))

print("subtraction")
print(tensor1 - 10)
print(torch.sub(tensor1,2))

print("element multiplication")
print(tensor1 * 10)
print(torch.mul(tensor1,10))
print(tensor1 * tensor1)

print("division")
print(tensor1 / 2)
print(tensor1.div(2))


print("matrix multiplication")
print(tensor1 @ tensor1)
print(torch.matmul(tensor1, tensor1))

```

```python
tensor_A = torch.tensor([[1,2],[3,4],[5,6]]) # 3,2
tensor_B = torch.tensor([[7,10],[8,11],[9,12]]) # 3,2

# transpose
print(torch.mm(tensor_A, tensor_B.T)) # alias of matmul
print(torch.mm(tensor_A.T, tensor_B)) # alias of matmul
```

## tensor reshape

```python
x = torch.arange(1,10)
print(x,x.shape)
x_reshaped = x.reshape(3,3)
print(x_reshaped)
z = x.view(9,1)
print(z.shape,x.shape)
z[0,:] = 5
print(x) # z and x use the same memory
```

```python
# stack
x_cat = torch.cat([x,x,x,x],dim = 0) # don't add dimension
print(x_cat)
x_stacked = torch.stack([x,x,x,x],dim=0) # add dimension
print(x_stacked)
```

```python
# squeeze: Returns a tensor with all specified dimensions of input of size 1 removed.
x_zeros = torch.zeros(2,1,2,1,2)
print(x_zeros,x_zeros.shape)
print(x_zeros.squeeze().shape)

print(x_cat.shape)

print(x_cat.unsqueeze(dim = 0).shape)
print(x_cat.unsqueeze(dim = 1).shape)
```

```python
#permute
random_image = torch.rand(size = (224,512,3)) #  height, width,color channels(R,G,B)
random_image_tensor = torch.permute(random_image,(2,0,1))
print(random_image_tensor.shape)
```

# tensor aggragation

```python
x = torch.arange(0,100,10)
print(torch.min(x), "," , x.min())
print(torch.max(x), "," , x.max())
print(torch.argmin(x), "," , x.argmin())
print(torch.argmax(x), "," , x.argmax())
print(torch.mean(x.type(torch.float32)), "," , x.type(torch.float32).mean())# requires float or complex datatype
print(torch.sum(x), "," , x.sum())
```

# Numpy to Pytorch tensor

```python
import numpy as np
# numpy to tensor
array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array)
print(array,array.dtype, tensor) # array is float64
tensor = torch.from_numpy(array).type(torch.float32)
print(tensor.dtype) # change to float32
# change array won'r impact tensor
```

```python
# tensor to numpy
tensor = torch.ones(7)
array = tensor.numpy()
print(array,array.dtype, tensor) # array is float32
```
