```python
#from numba import jit,njit
import numpy as np
```

# Without numba

```python
def numpy_features(matrix:np.array) -> None:
    cos_trace = 0.0
    for i in range(matrix.shape[0]):
        cos_trace += np.cos(matrix[i,i])
    matrix = matrix + cos_trace
```

```python
x = np.arange(1000000).reshape(1000,1000)
```

```python
%time
numpy_features(x)
```

# With numba

```python
#@jit(nopython = True)
#@njit
def numpy_features(matrix:np.array) -> None:
    cos_trace = 0.0
    for i in range(matrix.shape[0]):
        cos_trace += np.cos(matrix[i,i])
    matrix = matrix + cos_trace
```

```python
%time
numpy_features(x)
```
