# import package and config generation

```python
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
generation_config = {
    "temperature": 0.2,
    "top_p": 0.5,
    "top_k": 5,
    #"max_output_tokens": 2048, 
    "candidate_count":1,
    "stop_sequences":["END"],
    #"presence_penalty":0.5, # gemini-2.5-flash don't support
    #"frequency_penalty":0.3,# gemini-2.5-flash don't support
    "response_mime_type":"text/plain" # or "application/json"
}
```

# set up a structure prompt

```python
question = "tell me how temperature impact the answer in genai API config"
# pre set prompt
prompt_parts = [
        "Please limit your response to 5,000 words.",
    # "Role: You are an expert technical architect. Your goal is to provide exhaustive, deep-dive explanations with high technical density.", 
    # "Tone and Style:",
    # "1. Zero Pedantry: Strictly avoid condescending phrases like \"Many people overlook,\" \"The difference between experts and beginners,\" or \"It's simpler than it looks.\" ",
    # "2. Comprehensive Depth: For every concept, explain the \"Why\" and \"How\" in great detail. Break down complex mechanisms into logical components. Use structured analysis (e.g., pros/cons, trade-offs, internal workflows).",
    # "3. Objective Directness: Maintain a professional, neutral, and factual tone. No conversational fluff.",
    # "Output Structure:",
    # "1. Deep-Dive Explanation (80\\% of response): Provide a long-form, detailed technical breakdown. Cover edge cases, underlying architecture, and performance implications. ",
    # "2. Next Step (10\\% of response): In 1-2 concise sentences, provide the single most critical, actionable industry-standard next step.",
    # "3. Reflective Challenge (10\\% of response): Conclude with ONE brief, high-impact question regarding fundamental architectural trade-offs.",
    # "my question is: ", 
    {question}
]
```

# Send general request to Gemini

```python
client = genai.Client()
model = "gemini-2.5-flash"
response = client.models.generate_content(
    model=model, 
    contents=prompt_parts,
    config=generation_config
)
print(response.text)
#print(response.prompt_feedback)
#print(response.candidates)
```

# check token usage

```python
# check token usage
if response.usage_metadata:
    print("\n--- Token Usage Metadata ---")
    print(f"Prompt Token Count: {response.usage_metadata.prompt_token_count}")
    print(f"Candidates Token Count (Generated Output): {response.usage_metadata.candidates_token_count}")
    print(f"Total Token Count: {response.usage_metadata.total_token_count}")
else:
    print("\nNo usage metadata found in the response. This can happen with streaming or certain error states.")

# check stop reason if max_output_tokens is set:
print(response.candidates[0].finish_reason)

# check token count before getting response
token_count = client.models.count_tokens(contents="how many tokens?", model = model)
print("total tokens:",token_count.total_tokens)
print("cached content tokens:",token_count.cached_content_token_count)
print("sdk_http_response:",token_count.sdk_http_response)
```

# Streaming Responses

```python
# save outputs from stream generate
full_text = [] 
final_usage = None
response2 = client.models.generate_content_stream(
    model="gemini-2.5-flash", contents="I got this error when using genai API: AttributeError: 'generator' object has no attribute 'usage_metadata' the sode is: usage = response.usage_metadata"
)

for chunk in response2:
    print(chunk.text, end="", flush=True)
    full_text.append(chunk.text) # save outputs
    if chunk.usage_metadata:
        final_usage = chunk.usage_metadata

# check usage
print("\nUsage Metadata:", final_usage)
print(f"Input Tokens: {final_usage.prompt_token_count}")
print(f"Output Tokens: {final_usage.candidates_token_count}")
print(f"Total Tokens: {final_usage.total_token_count}")

# combine all outputs
final_answer = "".join(full_text)
```

```python
# print stream answer
print("\n--- 以下是缓存的完整内容 ---")
print(final_answer)
```
