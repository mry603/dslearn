# Randomness
## Generators

```python
# import random package
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
# create a generator
rng = random.default_rng()
# add seeds
rng = random.default_rng(42)
print(type(rng))
```

## Random Integers (inclusive and exclusive)

```python
# create random integer
rng.integers(0,10,size = 10) # generate ten random ints between [0,10)
rng.integers(0,10,size = 10, endpoint=True) # between [0,10]

# ramdom bits
rng.integers(2,size = 64)
```

## Random Function

```python
# random floats between 0 and 1
uniform_float = rng.random(50)
print(uniform_float)

# drawing random
x = rng.random(300)
y = rng.random(300)
plt.scatter(x,y)
plt.show()
```

```python
# shuffle
array = np.array([4,8,15,16,23,42])

rng.shuffle(array)
print("Shuffled array: ", array)
```

```python
# Choice
three_elements_from_array = rng.choice(array, size = 3)
print("choice:", three_elements_from_array)

print("-----------------")
newchoice = rng.choice(array,size = (2,2),replace = False) #size can be multi dimension, no replacement
print(newchoice)
```

# Statistics
## Normal Distribution and its statistic

```python
# Normal distribution
heights = rng.normal(1.75,0.15,100) # mean, stdev,size
std_normal = rng.standard_normal(100) # mean = 0, stdev =1,size = 100
plt.hist(std_normal)
plt.show()

print("Mean:", np.mean(std_normal))
print("Standard deviation: ", np.std(std_normal))
print("Variance:", np.var(std_normal))
print("Median:", np.median(std_normal))
```

## unique Value and modes

```python
#generate int normal dist
data_float = rng.normal(5,2, size=20)
data_int = np.round(data_float).astype(int)
plt.hist(data_int)
plt.show()
```

```python
# Remove duplicates and sort
values, counts=np.unique(data_int,return_counts=True)
values, index=np.unique(data_int,return_index=True)
values, inverse = np.unique(data_int,return_inverse=True)
print("original array:", data_int)
print("find unique array: ",values )
print("find counts: ",counts )
print("find unique index: ",index )
print("find unique inverse: ",inverse )

print("mode:",values[np.argmax(counts)]) # this way only acan find one mode value
print("-------------------------------")
max_count = np.max(counts)
modes = values[counts == max_count]
print("modes:",modes) # this way can find all modes

# draw a bar plot
plt.bar(values, counts)
plt.show()
```
