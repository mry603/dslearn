```python
from google import genai
```

```python
client = genai.Client()
model = "gemini-2.5-flash"
chat = client.chats.create(
    model=model,
    config={
        "system_instruction":"you are a professional psychological counselor",
    }
)
response = chat.send_message("Hello. I feel bad not finding jobs for long time. with current situation of AI technology exploding, I don't really think companies need to hire some engineers. However I still need job to earn money.")
print(response.text)
```

```python
# save history to a txt file
history = chat.get_history()

with open("newjob_find_consultant.txt", "w", encoding="utf-8") as f:
    for content in history:
        role = content.role
        text = content.parts[0].text if content.parts else ""
        
        f.write(f"【{role.upper()}】:\n{text}\n")
        f.write("-" * 30 + "\n") # split chat

print("save successful: chat_history.txt")
```

```python
# save as json for further input
import json

history = chat.get_history()
history_data = []

for content in history:
    history_data.append({
        "role": content.role,
        "parts": [{"text": p.text} for p in content.parts if p.text]
    })

with open("newjob_find_strategy.json", "w", encoding="utf-8") as f:
    json.dump(history_data, f, ensure_ascii=False, indent=4)
```

```python
import json

# 从本地文件读取
with open("newjob_find_strategy.json", "r", encoding="utf-8") as f:
    history_data = json.load(f)

# 打印一下检查数据结构
print(f"read {len(history_data)} history")
```

```python
# 假设你已经读取了刚才保存的 history_data
client = genai.Client()
model = "gemini-2.5-flash"
chat2 = client.chats.create(
    model=model,
    history=history_data, 
    #config=generation_config
)
response3 = chat2.send_message("yes you are right. but I feel bad when I talk about that. normally the listeners don't give me response.I feel they don't really care about what I was talking, or they feel my work useless.sometimes they say \"you should ... you should not... you know what you cannot do that...\" which really hurts me")
print(response3.text)
```

```python
# 获取 Token 使用情况
usage = response3.usage_metadata

print(f"--- Token 消耗统计 ---")
print(f"提示词 Token (Prompt Tokens): {usage.prompt_token_count}")
print(f"回答 Token (Candidates Tokens): {usage.candidates_token_count}")
print(f"总计 Token (Total Tokens): {usage.total_token_count}")
```
