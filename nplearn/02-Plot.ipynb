```python
import numpy as np
import matplotlib.pyplot as plt
```

# Universal Function

```python
vector = np.arange(1,5)
np.exp(vector)
```

```python
np.cos(np.pi*vector)
```

```python
np.log(np.e*vector)
```

# matplotlib plotting
## Function Plot

```python
x_value = np.arange(5)
y_value = x_value**2
plt.plot(x_value, y_value)
plt.show()
```

```python
x_value = np.linspace(0,np.pi,100)
y_value = np.cos(x_value)*2
plt.plot(x_value,y_value)
plt.xlabel("x")
plt.ylabel("$2*\\cos(x)$")
plt.show()
```

## Bar plot

```python
vector = np.array([5,1,12,0,3])
num_day = np.arange(1,len(vector)+1)
plt.bar(num_day,vector)
plt.xlabel("day")
plt.ylabel("rain in mm")
plt.title("Rain")
plt.show()
```

## Scatter Plot

```python
x_value = np.array([1,2,3,3.5,5,6.9,7,9])
y_values = np.array([1.5,2,2.5,4,5,6,7.2,8.5])
plt.scatter(x_value, y_values)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Experiment")
plt.show()
```

## Hist plot

```python
y_values = np.array([1,0,2,4,5,6,7,6,4,9,8,0,0,4,1,9])
plt.hist(y_values)
plt.show()
```
