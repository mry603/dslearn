```python
from google import genai
from pathlib import Path
import PIL.Image
```

# import picture url

```python
image_url = "cat.jpeg"
if not(img :=Path(image_url)).exists():
    raise FileNotFoundError(f"couldn't find images:{img}")
```

# Method 1 for Sending Images and Photos - send bytes

```python
image_part = genai.types.Part.from_bytes(
    data=Path(image_url).read_bytes(),
    mime_type="image/jpeg"
)

prompt_parts = [
    "what animal is in this picture?",
    image_part
]
```

```python
client = genai.Client()
model = "gemini-2.5-flash"
response = client.models.generate_content(
    model=model, 
    contents=prompt_parts,
    #config=generation_config
)
print(response.text)
```
