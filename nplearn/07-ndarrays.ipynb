# Numpy ndarrays
## making ndarrys

```python
import numpy as np
```

```python
# 3D list
d3_list = [[[1,2],[3,4]],[[5,6],[7,8]]]

# convert into 3D array
d3_array = np.array(d3_list)
print(d3_array)

# convert back to list
d3_array_list = d3_array.tolist()
print(d3_array_list)
```

```python
# indexing
print(d3_array[0,1,0])
d3_array[1,0,1] = 99
print(d3_array)
print(d3_array.shape)
```

```python
# 4D list
d4_array = np.zeros((3,3,2,2))
print(d4_array)
```

## slicing and aggregating

```python
# make an array of random integers
rng = np.random.default_rng()
array = rng.integers(0,10,size = 36).reshape(4,3,3)
print(array)
print(array[0,0:2,:])
print(array[2,:,[0,2]])
array[array >= 8] = 30
print(array)
```

```python
# aggregate functions
print(array.sum(axis = 0))
```

# application：colored images

```python
from skimage import data
import matplotlib.pyplot as plt

cat = data.chelsea()
plt.imshow(cat)
print(cat.shape)# the last axis is RGB color
```

```python
red_cat = cat.copy()
red_cat[:,:,[1,2]] = 0 # remove other color
red_cat = red_cat[:,:-1,:] # delete the last column
print(red_cat.shape)
plt.imshow(red_cat)
```

```python
copycat = cat.copy()
print(copycat.shape)
copycat[copycat[:,:,0]>160] = [0,255,0]
plt.imshow(copycat.transpose((1,0,2))) # rows and columns exchanged, keep colors at the last
```

# strides

```python
my_array = np.arange(12).reshape(4,3)
print("datatype:", my_array.dtype) # 8bytes
print("strides:",my_array.strides) 
print(my_array.flatten())
my_array_t = my_array.transpose()
print("datatype:", my_array_t.dtype) # 8bytes
print("strides:",my_array_t.strides) 
print(my_array_t.flatten())

my_array_odd = my_array[::2,::2]
print("datatype:", my_array_odd.dtype) # 8bytes
print("strides:",my_array_odd.strides) 
print(my_array_odd.flatten())
print(my_array_odd)
```
